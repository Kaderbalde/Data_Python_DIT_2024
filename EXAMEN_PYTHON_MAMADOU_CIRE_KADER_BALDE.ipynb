```python
#impoertation des packages
import numpy as np
import matplotlib.pyplot as plt

# Définition de la fonction x^5
def f(x):
    return x ** 5

# Création de valeurs x pour la courbe
x = np.linspace(0, 1, 400)

# Création de valeurs y en utilisant la fonction f(x)
y = f(x)

# Tracer la courbe
plt.plot(x, y, label='$y = x^5$', color='blue')

# Tracer les lignes du carré
plt.plot([0, 1], [0, 0], color='red') # ligne basse
plt.plot([1, 1], [0, 1], color='red') # ligne droite
plt.plot([1, 0], [1, 1], color='red') # ligne haute
plt.plot([0, 0], [1, 0], color='red') # ligne gauche

# Paramètres du graphique
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbe $y = x^5$ et carré délimité par x=0, x=1, y=0, y=1')
plt.grid(True)
plt.axis('equal')
plt.legend()

# Affichage du graphique
plt.show()
```

```python
#Importation du package
import numpy as np
import matplotlib.pyplot as plt

class Curve:
    def __init__(self, start, stop, nbr_points=5432):
        self.start = start
        self.stop = stop
        self.nbr_points = nbr_points

    def _f(self, x):
        return x ** 5

    def plot_curve(self):
        x_values = np.linspace(self.start, self.stop, self.nbr_points)
        y_values = self._f(x_values)
        plt.plot(x_values, y_values, label='$y = x^5$', color='blue')

    def above_or_below(self, x, y):
        if y > self._f(x):
            return "above"
        elif y < self._f(x):
            return "below"
        else:
            return "on curve"

    def plot_points(self):
        x_values = np.linspace(self.start, self.stop, self.nbr_points)
        y_values_curve = self._f(x_values)
        y_values_points = np.random.uniform(0, 1, self.nbr_points) # génère des y aléatoires dans [0, 1]

        areas_blue = []
        areas_green = []

        for i in range(self.nbr_points):
            if y_values_points[i] > y_values_curve[i]:
                plt.plot(x_values[i], y_values_points[i], 'bx') # point au dessus de la courbe
                areas_blue.append(y_values_points[i] - y_values_curve[i])
            else:
                plt.plot(x_values[i], y_values_points[i], 'go') # point en dessous de la courbe
                areas_green.append(y_values_curve[i] - y_values_points[i])

        area_blue = sum(areas_blue) * (self.stop - self.start) / self.nbr_points
        area_green = sum(areas_green) * (self.stop - self.start) / self.nbr_points

        return area_blue, area_green

#Test d'utilisation
curve = Curve(0, 1)
curve.plot_curve()
area_blue, area_green = curve.plot_points()
print("Surface en bleu :", area_blue)
print("Surface en vert :", area_green)

# Paramètres du graphique
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbe $y = x^5$ avec points générés')
plt.grid(True)
plt.legend()

# Affichage du graphique
plt.show()
#pour calculer la surface en bleu et celle en vert, on fait ls sommes des ecarts  entre les points de la courbe.
```
